```python
import cv2
import numpy as np
import pandas as pd
```

```python
img = cv2.imread('galaxy_image.jpg')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print(f"Image shape: {img_rgb.shape}")
```

```python
r_values = img_rgb[:, :, 0]  # Red channel
g_values = img_rgb[:, :, 1]  # Green channel
b_values = img_rgb[:, :, 2]  # Blue channel
```

```python
pixel_data = img_rgb.reshape(-1, 3)  # Flatten and retain RGB channels
df = pd.DataFrame(pixel_data, columns=['Red', 'Green', 'Blue'])
```

```python
df.to_csv('pixel_data.csv', index=False)
print("Pixel data saved to 'pixel_data.csv'")
```
